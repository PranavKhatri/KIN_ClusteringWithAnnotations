```python
def uniprot_to_RNAseq(edges_no_weights_file = 'data/Full_Kinome_Network_Compiled_no_header.txt',
                      weighted_edges_file='data/Full_Kinome_Network_Compiled_weighted_pathways.txt',
                      alias_file = 'data/KINASESmasterlist_w_Aliases.xlsx'):
    
    edges_no_weights = pd.read_csv(edges_no_weights_file, sep='\t', header=None)
    weighted_edges = pd.read_csv(weighted_edges_file,  sep='\t', header=None)
    aliases = pd.read_excel(alias_file)
    
    alias_dict = aliases[['Uniprot Protein','RNAseq Gene']].set_index('Uniprot Protein').to_dict()['RNAseq Gene']
    #print(alias_dict)
    cols_no_weights = edges_no_weights.columns
    weighted_cols = weighted_edges.columns
    
    for i in range(len(cols_no_weights)):
        edges_no_weights[cols_no_weights[i]] = edges_no_weights[cols_no_weights[i]].map(alias_dict)
    
    edges_no_weights.to_csv('data/KIN_edges_no_weights.txt', header=False, index=False)
    
    for i in range(len(weighted_cols)-1):
        weighted_edges[weighted_cols[i]] = weighted_edges[weighted_cols[i]].map(alias_dict)
        
    weighted_edges.to_csv('data/KIN_weighted_edges.txt', header=False, index=False)
        
```

```python
uniprot_to_RNAseq()
```
